```python
#Q15 - Write a python using conditional statements.
#Check wether it is armstrong or not

num = int(input("Enter a number: "))

sum_cube = 0

temp = num
while temp > 0:
   digit = temp % 10
   sum_cube += digit ** len(str(num))
   temp //= 10

if num == sum_cube:
   print(num,"is an Armstrong number")
else:
   print(num,"is not an Armstrong number")
```
