***10. Fuctions***

```python
Aditya_marks = [45, 55, 65, 35, 50]
Virendra_marks = [89, 88, 67, 50, 75]
Vivaan_marks = [55, 75, 85, 90, 70]
Naman_marks = [90, 55, 60, 75, 80]
Jayesh_marks = [99, 98, 97, 96, 95]

sum = 0
for i in Aditya_marks:
  sum = sum + i
print("Adtiya: ", sum)

sum = 0
for i in Vivaan_marks:
  sum = sum + i
print("Vivaan: ", sum)

sum = 0
for i in Virendra_marks:
  sum = sum + i
print("Virendra: ", sum)

sum = 0
for i in Naman_marks:
  sum = sum + i
print("Naman: ", sum)

sum = 0
for i in Jayesh_marks:
  sum = sum + i
print("Jayesh: ", sum)

```

```python
def marks(student):
  sum1 = 0
  for i in student:
    sum1 = sum1 + i
  return sum1
```

```python
Aditya = marks(Aditya_marks)
Vivaan = marks(Vivaan_marks)
Virendra = marks(Virendra_marks)
Naman = marks(Naman_marks)
Jayesh = marks(Jayesh_marks)

print("Aditya: ", Aditya)
print("Vivaan: ", Vivaan)
print("Virendra: ", Virendra)
print("Naman: ", Naman)
print("Jayesh: ", Jayesh)
```
