```python
# Display numbers from a light using a loop
# 1. The number must be divisble by 5
# 2. The number should be greater than 150
# 3. If the number is greater than 500 then stop the loop
numbers = [12, 75, 150, 180, 145, 525, 50, 877]
for i in numbers:
    if i > 150:
        if i > 500:
            break
        if i % 5 == 0:
                print(i)
```

```python
# Count the total number of digits in a number
num = int(input("Enter a number: "))
count = 0
while num != 0:
    num //= 10
    count += 1
print("Total digits are: ", count)
```

```python
# print the pattern
rows = 5
column = 5
for i in range(0, rows + 1):
    for j in range(column-i, 0, -1):
      print(j, end = ' ')
    print()
```

```python
# Fibonacci series
n = int(input("Enter number of terms: "))
a, b = 0, 1
print("Fibonacci series: ")
for i in range(n):
    print(a, end=" ")
    a, b = b, a + b
```

```python
# Print factorial of a given number
num = 5
fac = 1
if num < 0:
  print("Factorial does not exist for negative numbers")
elif num == 0:
  print("The factorial of 0 is 1")
else:
  for i in range(1, num + 1):
    fac *= i
  print(fac)
```

```python
# Reverse a integer number
# given = 765432, expected output:234567
num = 765432
rev = 0
print("Given number is: ",num)
while num>0:
  rem = num % 10
  rev = (rev*10) + rem
  num = num // 10
print("Reversed number is: ",rev)
```
